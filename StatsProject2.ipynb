## Additional Statistical Analysis


```python

from scipy.stats import ttest_ind, chi2_contingency

# Example: T-test for sales in different regions
sales_east = df[df['Region'] == 'East']['Sales']
sales_west = df[df['Region'] == 'West']['Sales']



t_stat, p_value = ttest_ind(sales_east, sales_west, equal_var=False)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")







```

```python
# Example: Chi-square test for independence between Ship Mode and Customer Segment

contingency_table = pd.crosstab(df['Ship Mode'], df['Customer Segment'])

print(contingency_table)

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", expected)


```

```python

# Plotting sales trends over time

df2['Order Date'] = pd.to_datetime(df2['Order Date'])

sales_trend = df2.resample('M', on='Order Date')['Sales'].sum()

plt.figure(figsize=(8, 6))
plt.plot(sales_trend.index, sales_trend.values, marker='o')
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()









```

```python

# Plotting profit trends over time

df2['Order Date'] = pd.to_datetime(df2['Order Date'])
monthly_profit = df2.resample('M', on='Order Date')['Profit'].sum()
plt.figure(figsize=(8, 6))
plt.plot(monthly_profit.index, monthly_profit.values, marker='o', color='green')
plt.title('Profit Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()




```

```python

# Identifying top-performing product categories based on total sales and profit


category_performance = df2.groupby('Product Category').agg({
    'Sales': 'sum',
    'Profit': 'sum'
})

top_categories_sales = category_performance.sort_values(by='Sales', ascending=False)
top_categories_profit = category_performance.sort_values(by='Profit', ascending=False)
#sales
plt.figure(figsize=(6, 4))
sns.barplot(x=top_categories_sales.index, y=top_categories_sales['Sales'], palette='Blues_d')
plt.title('Top-Performing Product Categories by Total Sales')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

#profit
plt.figure(figsize=(6, 4))
sns.barplot(x=top_categories_profit.index, y=top_categories_profit['Profit'], palette='Greens_d')
plt.title('Top-Performing Product Categories by Total Profit')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

















```
